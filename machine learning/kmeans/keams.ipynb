```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import calinski_harabasz_score

x, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1, -1], [0, 0], [1, 1], [2, 2]], cluster_std=[0.4, 0.2, 0.3, 0.4], random_state=23)
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()

estimator = KMeans(n_clusters=4, random_state=23)  # n_clusters: Valid values are 2 to n_samples - 1 (inclusive)
plt.scatter(x[:, 0], x[:, 1], c=estimator.fit_predict(x))
plt.show()

print(f'Calinski Harabasz score: {calinski_harabasz_score(x, estimator.labels_)}')
```

```python
# sse (sum of squared errors) 评估系数  肘部法则

sse_lst = []

x, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [0,0], [1,1], [2,2]], cluster_std=[0.4, 0.2, 0.3, 0.4], random_state=23)

for k in range(1, 100):
    estimator = KMeans(n_clusters=k, max_iter=100, random_state=23)
    estimator.fit(x)
    sse_lst.append(estimator.inertia_)

plt.figure(figsize=(20, 10))
plt.title('SSE curve')
plt.xticks(range(0, 100, 3))
plt.xlabel('K')
plt.ylabel('SSE value')
plt.grid()
plt.plot(range(1, 100), sse_lst, marker='o')
plt.show()
```

```python
# sc (silhouette coefficient) 评估系数

from sklearn.metrics import silhouette_score

sc_lst = []

x, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [0,0], [1,1], [2,2]], cluster_std=[0.4, 0.2, 0.3, 0.4], random_state=23)

for k in range(2, 100):
    estimator = KMeans(n_clusters=k, max_iter=100, random_state=23)
    estimator.fit(x)
    sc_lst.append(silhouette_score(x, estimator.labels_))

plt.figure(figsize=(20, 10))
plt.title('Silhouette Coefficient curve')
plt.xticks(range(0, 100, 3))
plt.xlabel('K')
plt.ylabel('Silhouette Coefficient')
plt.grid()
plt.plot(range(2, 100), sc_lst, marker='o')
plt.show()
```

```python
# ch (calinski-harabasz index) 评估系数

from sklearn.metrics import silhouette_score

ch_lst = []

x, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [0,0], [1,1], [2,2]], cluster_std=[0.4, 0.2, 0.3, 0.4], random_state=23)

for k in range(2, 100):
    estimator = KMeans(n_clusters=k, max_iter=100, random_state=23)
    estimator.fit(x)
    ch_lst.append(calinski_harabasz_score(x, estimator.labels_))

plt.figure(figsize=(20, 10))
plt.title('Calinski-Harabasz index curve')
plt.xticks(range(0, 100, 3))
plt.xlabel('K')
plt.ylabel('Calinski-Harabasz Coefficient')
plt.grid()
plt.plot(range(2, 100), ch_lst, marker='o')
plt.show()
```
