```python
from sklearn.linear_model import LinearRegression

x_train = [[160], [166], [172], [174], [180]]       # 训练集的特征
y_train = [56.3, 60.6, 65.1, 68.5, 75]              # 训练集的标签
x_test = [[176]]                                    # 测试集的特征

estimator = LinearRegression()
estimator.fit(x_train, y_train)

print(f"weight: {estimator.coef_}")
print(f"bias: {estimator.intercept_}")

y_pred = estimator.predict(x_test)

cal = estimator.coef_ * x_test[0][0] + estimator.intercept_

print(f"y_pred: {y_pred}")
print(f'self cal: {cal}')
```

求解模型参数: norm equation; gradient descent
norm equation: 
    求解的时候需要求矩阵可逆，所以可用gram matrix使其成方阵，若可逆，则可等式两边求逆
    运算量容易过大；若矩阵没有逆，则无法计算
    pandas中，一行数据是通过列的形式展示；多行数据是通过行的形式展示；所以这里会有矩阵乘法维度的公式调整（了解）

gradient descent:
    theta = theta - learning rate * partial loss function(theta)
    category: full gradient descent, stochastic gradient descent, mini-batch gradient descent, stochastic average gradient descent 
    best for choice: mini-batch gd


loss function:
    mean squared error, root mean squared error (对outlier敏感)
    mean absolute error
