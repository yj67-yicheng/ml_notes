confusion matrix:
    true positive, false positive
    true negative, false negative
    (上面的术语都是针对预测结果说的)

    precision: 查准，true positive 在 pred positive 中的占比，按col看
    recall：查全，true positive 在 label positive 中的占比, 按row看
    accuracy
    f1：结合precision, recall


```python
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# 1. 定义变量, 记录: 样本数据
y_gt = ['恶性', '恶性', '恶性', '恶性', '恶性', '恶性',     '良性', '良性', '良性', '良性']

# 2. 定义变量, 记录: 模型A的预测结果
y_pred_A = ['恶性', '恶性', '恶性', '良性', '良性', '良性',    '良性', '良性', '良性', '良性']

# 3. 定义变量, 记录: 模型B的预测结果
y_pred_B = ['恶性', '恶性', '恶性', '恶性', '恶性', '恶性',     '良性', '恶性', '恶性', '恶性']

# 4. 用标签标记 正例, 反例.
label = ['恶性', '良性']  # confusion matrix会自动按照index顺序排列, 这里恶性是正例, 良性是反例.
df_label = ['恶性(正例)', '良性(反例)']

cm_A = confusion_matrix(y_gt, y_pred_A, labels=label)
cm_B = confusion_matrix(y_gt, y_pred_B, labels=label)
df_A = pd.DataFrame(cm_A, index=df_label, columns=df_label)
df_B = pd.DataFrame(cm_B, index=df_label, columns=df_label)

print(f'confusion matrix A: \n {df_A}')
print(f'confusion matrix B: \n {df_B}')
print()

print(f'precision A: {precision_score(y_gt, y_pred_A, pos_label="恶性")}')  # pos_label参数指定哪个标签是正例, 这里恶性是正例.
print(f'precision B: {precision_score(y_gt, y_pred_B, pos_label="恶性")}')
print()
print(f'recall A: {recall_score(y_gt, y_pred_A, pos_label="恶性")}')
print(f'recall B: {recall_score(y_gt, y_pred_B, pos_label="恶性")}')
print()
print(f'f1 score A: {f1_score(y_gt, y_pred_A, pos_label="恶性")}')
print(f'f1 score B: {f1_score(y_gt, y_pred_B, pos_label="恶性")}')

```

Receiver Operating Characteristic curve (ROC)
    计算时有隐藏阈值 t \in [0, 1]
    y轴：TPR, true positive rate 正样本中被预测为正样本的概率 = TP / (TP + FN)，伪正率
    x轴：FPR, false positive rate, 负样本中被预测为正样本的概率 = FP / (FP + TN)，真正率
    都是按行看
    有4个特殊点：(0, 1), (0, 0), (1, 0), (1, 1)

Area under the ROC curve (AUC)
y = x, auc = 0.5 -> random guess; ROC向左上角鼓，AUC大，classifier效果好
一般用于二分类
