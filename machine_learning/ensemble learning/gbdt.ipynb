```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

df = pd.read_csv('../data/train.csv')
# df.info()

x = df[['Pclass', 'Sex', 'Age']].copy()
y = df['Survived'].copy()

x['Age'] = x['Age'].fillna(x['Age'].mean())
x = pd.get_dummies(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=23)

estimator = DecisionTreeClassifier(random_state=23)
estimator.fit(x_train, y_train)
print(f'Decision Tree Classifier Score: {estimator.score(x_test, y_test)}')

estimator2 = GradientBoostingClassifier(random_state=23)
estimator2.fit(x_train, y_train)
print(f'Graident Boosting Classifier Score: {estimator2.score(x_test, y_test)}')

param = {'n_estimators': [50, 60], 'learning_rate': [0.3, 0.5], 'max_depth': [3, 5]}
estimator3 = GridSearchCV(estimator2, param, cv=2)
estimator3.fit(x_train, y_train)
print(f'Gradient Boosting Classifier with GridSearch Score: {estimator3.score(x_test, y_test)}')

print(f'After GridSearch, Acc: {estimator3.best_score_}')
print(f'After GridSearch, Best Params: {estimator3.best_params_}')
```
