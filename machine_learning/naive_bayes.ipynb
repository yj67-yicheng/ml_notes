```python
"""
概述:
        贝叶斯: 仅仅依赖 概率 就可以进行分类的 一种机器学习算法.
        朴素:   不考虑特征之间的关联性, 即: 特征间都是相互独立的.
            原始:  P(AB) = P(A) * P(B|A) = P(B) * P(A|B)
            加入朴素后: P(AB) = P(A) * P(B)
细节:
    因为我们分词要用到 jieba分词器, 记得先装一下, 例如: pip install jieba
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import jieba
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

df = pd.read_csv('./data/书籍评价.csv', encoding='gbk')
df.info()
print()

df['labels'] = np.where(df['评价'] == '好评', 1, 0)
print(df)
print()

y = df['labels']
comment_lst = [', '.join(jieba.lcut(line)) for line in df['内容']]
print(comment_lst)
print()

with open('./data/stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.readlines()
    stop_words = [line.strip() for line in stop_words]
    stop_words = list(set(stop_words))

    print(stop_words)
print()

transfer = CountVectorizer(stop_words=stop_words)
x = transfer.fit_transform(comment_lst).toarray()  # type: ignore
print(x)
print()

print(transfer.get_feature_names_out())
print(len(transfer.get_feature_names_out()))
print()

x_train = x[:10]
y_train = y[:10]
x_test = x[10:]
y_test = y[10:]

estimator = MultinomialNB()
estimator.fit(x_train, y_train)
print(estimator.score(x_test, y_test))
```
