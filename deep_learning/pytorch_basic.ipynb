```python
import torch
from torch.utils.data import TensorDataset, DataLoader
from torch import nn
from torch import optim
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

def create_dataset():
    x, y, coef = make_regression(
        n_samples=100, n_features=1,
        noise=10, coef=True, bias=14.5,
        random_state=3,
    )

    x = torch.tensor(x, dtype=torch.float32)
    y = torch.tensor(y, dtype=torch.float32)

    return x, y, coef

def train(x, y, coef):
    dataset = TensorDataset(x, y)
    dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
    linear = nn.Linear(1, 1)
    criterion = nn.MSELoss()
    optimizer = optim.SGD(linear.parameters(), lr=0.01)
    epochs, loss_lst, tot_loss, tot_samples = 100, [], 0.0, 0
    for epoch in range(epochs):
        for x_train, y_train in dataloader:
            y_pred = linear(x_train)
            loss = criterion(y_pred, y_train.reshape(-1, 1))
            tot_loss += loss.item()
            tot_samples += 1
            optimizer.zero_grad()
            loss.sum().backward()
            optimizer.step()

        loss_lst.append(tot_loss / tot_samples)
        print(f'epoch: {epoch + 1}, avg loss: {tot_loss / tot_samples}')
    
    print(f'model coef: {linear.weight.item()}, model bias: {linear.bias.item()}')

    plt.plot(range(1, epochs + 1), loss_lst)
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.show()

    plt.scatter(x, y)
    y_pred = torch.tensor(data=[v * linear.weight.item() + linear.bias.item() for v in x])
    y_true = torch.tensor(data=[v * coef + 14.5 for v in x])
    plt.plot(x, y_true, label='real value', color='red')
    plt.plot(x, y_pred, label='pred value', color='blue')
    plt.legend()
    plt.grid()
    plt.show()

    
```

```python
x, y, coef = create_dataset()
train(x, y, coef)
```

```python
import torch
import torch.nn as nn
from torchsummary import summary

class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(3, 3)
        self.linear2 = nn.Linear(3, 2)
        self.output = nn.Linear(2, 2)

        nn.init.xavier_normal_(self.linear1.weight)
        nn.init.zeros_(self.linear1.bias)

        nn.init.kaiming_normal_(self.linear2.weight)
        nn.init.zeros_(self.linear2.bias)

    def forward(self, input):
        res = torch.sigmoid(self.linear1(input))
        res = torch.relu(self.linear2(res))
        res = torch.softmax(self.output(res), dim=-1)

        return res
    

```

```python
mlp = MLP()
data = torch.randn(size=(5, 3))
res = mlp(data)
print(f'res: {res}')
summary(mlp, input_size=(5, 3))

for name, param in mlp.named_parameters():
    print(f'name: {name}')
    print(f'param: {param}')
```

```python
def dm02(beta=0.9):    
    torch.manual_seed(0)
    temp = torch.randn(size=(30,)) * 10
    print(temp)

    ewa = []
    for idx, t in enumerate(temp):
        if idx == 0: 
            ewa.append(t)
            continue
        new_t = ewa[idx - 1] * beta + (1 - beta) * t
        ewa.append(new_t)

    days = torch.arange(1, 30 + 1)
    plt.plot(days, ewa, color='r')
    plt.scatter(days, temp, color='b')
    plt.show()

dm02()
    
```
